1º Etapa - Combinando arquivos CSV

```python
#Definindo diretorio
os.chdir("/home/brunodangelosilva/Documentos/IFSP/BIG DATA/1º Trabalho/arquivos_brutos")
```

```python
#Definitido o tipo de extensão que será lido
extension = 'csv'
csv_temp = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
# Inicializar DataFrame vazio
csv_final = pd.DataFrame()

# Loop para ler cada arquivo e concatenar
for i, f in enumerate(csv_temp):
     if i == 0:
         # Para o primeiro arquivo, ler o cabeçalho
         csv_final = pd.read_csv(f, on_bad_lines='skip', encoding='latin-1', delimiter=';', low_memory=False)
     else:
         # Para os demais arquivos, ignorar o cabeçalho
         csv_atual = pd.read_csv(f, on_bad_lines='skip', skiprows=1, encoding='latin-1', delimiter=';', low_memory=False)
         csv_final = pd.concat([csv_final, csv_atual])

#Exporta para csv
csv_final.to_csv( "csv_final.csv", index=False, encoding='latin-1')
```
