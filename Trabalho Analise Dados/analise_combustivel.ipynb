```python
estado_sp = df['Estado - Sigla'].str.contains('SP', case=False, na=False)

# Filtra o DataFrame para incluir apenas as linhas onde 'nm_ug_tabela_ug' contém "SAO PAULO"
sao_paulo_df = df[estado_sp]

# Exibe o DataFrame resultante
sao_paulo_df.head()
```

```python
# Agrupa o DataFrame por 'sg_orgao' e calcula a soma dos valores mensais de custo
resumo = df.groupby('Estado - Sigla')['Valor de Venda'].sum().reset_index()

# Ordena o DataFrame pelo valor mensal de custo em ordem decrescente
resumo = resumo.sort_values(by='Valor de Venda', ascending=False)

# Exibe o resumo ordenado
print(resumo)
```

```python
resumo = df.groupby('Estado - Sigla')['Valor de Venda'].sum().reset_index()

# Remove a formatação da coluna 'Valor de Venda' para a conversão
resumo['Valor de Venda'] = resumo['Valor de Venda'].str.replace('[$,]', '', regex=True).astype(float)

# Ordena o DataFrame pelo valor de venda em ordem decrescente
resumo = resumo.sort_values(by='Valor de Venda', ascending=False)

# Seleciona os 10 maiores valores
top_10 = resumo.head(10)

# Ajusta o tamanho da figura
plt.figure(figsize=(12, 8))

# Cria um gráfico de barras dos 10 maiores valores
bars = plt.barh(top_10['Estado - Sigla'], top_10['Valor de Venda'])
plt.xlabel('Soma dos Valores de Venda')
plt.ylabel('Estado - Sigla')
plt.title('Top 10 Estado - Sigla com Maior Soma de Valor de Venda')
plt.gca().invert_yaxis()  # Inverte o eixo y para ter os maiores no topo

# Adiciona os valores das barras dentro das barras
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, '{:.2f}'.format(width),
             ha='left', va='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```
