```python
# Carregue o conjunto de dados em um DataFrame
df = pd.read_csv("/home/brunodangelosilva/Documentos/IFSP/Trabalho Analise Dados/arquivos_brutos/ca-2023-01.csv", on_bad_lines='skip', encoding='latin-1', delimiter=';', low_memory=False)
```

```python
# Verifique se há valores ausentes
df.isnull().sum()
```

```python
# Agrupe os dados por estado e calcule a média do preço de venda
estado_precos = df.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
```

```python
# Ordene os estados por preço médio de venda do maior para o menor
estado_precos = estado_precos.sort_values(by='Valor de Venda', ascending=False)
```

```python
# Plote um gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(estado_precos['Estado - Sigla'], estado_precos['Valor de Venda'])
plt.xlabel('Estado')
plt.ylabel('Preço Médio de Venda (R$/litro)')
plt.title('Preço Médio de Venda de Combustíveis por Estado')
plt.xticks(rotation=45)
plt.show()
```
